```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
from google.colab import files

drive.mount('/content/drive')

df = pd.read_csv('/insurance.csv')

print("=== Первые 10 строк ===")
print(df.head(10))
print("\n=== Информация о данных ===")
print(df.info())
print("\n=== Статистическое описание ===")
print(df.describe())

print("\n=== Пропуски в данных ===")
print(df.isnull().sum())
print("\n=== Типы данных до преобразования ===")
print(df.dtypes)

df['sex'] = df['sex'].astype('category')
df['smoker'] = df['smoker'].astype('category')
df['region'] = df['region'].astype('category')

print("\n=== Типы данных после преобразования ===")
print(df.dtypes)

plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Распределение возраста')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(df['bmi'], df['charges'], alpha=0.5)
plt.title('Зависимость расходов от ИМТ')
plt.xlabel('ИМТ')
plt.ylabel('Расходы')
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(8, 6))
df['smoker'].value_counts().plot(kind='bar', color=['lightgreen', 'lightcoral'])
plt.title('Количество курящих и некурящих')
plt.xlabel('Курение')
plt.ylabel('Количество')
plt.xticks(rotation=0)
plt.grid(True, alpha=0.3)
plt.show()


plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black', alpha=0.8)
plt.title('Распределение возраста')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.grid(True, alpha=0.3)
plt.show()
```
